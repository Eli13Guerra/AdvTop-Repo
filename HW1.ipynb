```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # for plotting
import datetime
import seaborn as sns
from sklearn.impute import SimpleImputer

data = pd.read_csv('/Users/elimg/Documents/Jupyter Notebooks/Total Emissions Per Country (2000-2020).csv')
```

```python
x = np.array([i for i in range(21)], dtype='int')
brazil_fires = data.iloc[5994, 4:25].values
guate_fires = data.iloc[19339, 4:25].values
colombia_fires = data.iloc[10125, 4:25].values
```

```python
plt.scatter(x, brazil_fires, c='r')
plt.scatter(x, guate_fires, c='b')
plt.scatter(x, colombia_fires, c='g')
plt.xlabel("Years")
plt.ylabel("Emissions (CO2eq. kilotonnes)")
plt.xticks(x)
plt.title("Forest Fires in Latam in 20'")
plt.legend(['Brazil', 'Guatemala', 'Colombia'])
plt.show()
print ("Brazil has the largest CO2 emissions due to forest fires. This is because Brazil contain the largest share of the Amazon, which is the largest forest in America.")
print ("The CO2 emissions of Guatemala and Colombia due to forest fires are insignificant when compare with the ones of Brazil.")
```

```python
gua20 = data.iloc[19261:19490,24].values
gua_em = data.iloc[19261:19490,2].values
gua_list = data.iloc[19261:19490,1].values
gua20co2 = []
gualist = []    
i=0
while i<229:
    if gua_em[i]=="Direct emissions (N2O)":
        gua20co2.append(gua20[i])
        gualist.append(gua_list[i])
        i=i+1
    else: 
        i=i+1

#print(gua20co2)
#print(gualist)
#print(len(gua20co2),len(gualist))
```

```python
fig = plt.figure(figsize=(10,7))
plt.pie(gua20co2, labels = gualist)
plt.title("Direct emissions of N2O in 2020 in Guatemala")
plt.show()
print ("From the plot is visible that the use of fertilizers contributes to a large share of N2O emissions in 2020.")
print ("When comparing the emissions of manure applied and the emissions of manure left in pasture, the difference is significantly large. This means that most of the manure is left to decompose and not taken advantage of.")
```

```python
e_gua = data.iloc[19393, 4:25].values
e_salv = data.iloc[14490, 4:25].values
e_bel = data.iloc[4574, 4:25].values
e_cr = data.iloc[11046, 4:25].values
e_hon = data.iloc[20571, 4:25].values
e_nic = data.iloc[32357, 4:25].values
e_pan = data.iloc[34785, 4:25].values

plt.xlabel ("Year")
plt.xticks (x)
plt.ylabel ("Emission of CO2eq (AR5)")
plt.plot(x,e_gua)
plt.plot(x,e_salv)
plt.plot(x,e_bel)
plt.plot(x,e_cr)
plt.plot(x,e_hon)
plt.plot(x,e_nic)
plt.plot(x,e_pan)
plt.legend (["Guatemala","El Salvador","Belize","Costa Rica","Honduras","Nicaragua","Panama"])
plt.title ("Energy production CO2 emission in Central America in 20'")
plt.show()

print ("From the plot it it can be seen that all Central American countries had a sudden spike in CO2 emissions due to energy production in 2019. This is due to the fact that the largest consumption in Central American countries comes from the domestic sector and not the industrial sector, like in developed countries. The COVID-19 pandemic started in 2019, making people stay at home and increase their power demand and therefore, the power generation.")
```
